### Number and Arrays

$\begin{array}{cl}
    a & \text{A scalar (integer or real)}\\
    \va & \text{A vector}\\
    \mA & \text{A matrix}\\
    \tA & \text{A tensor}\\
    \mI_n & \text{Identity matrix with n rows and n columns}\\
    \mI & \text{Identity matrix with dimensionality implied by context}\\
    \ve^{(i)} & \text{Standard basis vector [0, ...,0 ,1 ,0, ..., 0] with a 1 at position i}\\
    \text{diag}(\va) & \text{A square, diagonal matrix with diagonal entries given by }\va\\
    \ra & \text{A scalar random variable}\\
    \rva & \text{A vector-valued random variable}\\
    \rmA & \text{A matrix-valued random variable}
\end{array}$

### Sets and Graphs

$\begin{array}{cl}
    \sA & \text{A set}\\
    \sR & \text{The set of real numbers} \\
    \{0, 1\} & \text{The set containing 0 and 1} \\
    \{0, 1, \dots, n \} & \text{The set of all integers between $0$ and $n$}\\
    [a, b] & \text{The real interval including $a$ and $b$}\\
    (a, b] & \text{The real interval excluding $a$ but including $b$}\\
    \sA \backslash \sB & \text{Set subtraction, i.e., the set containing the elements of $\sA$ that are not in $\sB$}\\
    \gG & \text{A graph}\\
    P_\gG(\ervx_i) & \text{The parents of $\ervx_i$ in $\gG$}
\end{array}$

### Indexing

$\begin{array}{cl}
    \eva_i & \text{Element $i$ of vector $\va$, with indexing starting at 1} \\
    \eva_{-i} & \text{All elements of vector $\va$ except for element $i$} \\
    \emA_{i,j} & \text{Element $i, j$ of matrix $\mA$} \\
    \mA_{i, :} & \text{Row $i$ of matrix $\mA$} \\
    \mA_{:, i} & \text{Column $i$ of matrix $\mA$} \\
    \etA_{i, j, k} & \text{Element $(i, j, k)$ of a 3-D tensor $\tA$}\\
    \tA_{:, :, i} & \text{2-D slice of a 3-D tensor}\\
    \erva_i & \text{Element $i$ of the random vector $\rva$} \\
\end{array}$

### Linear Algebra Operations

$\begin{array}{cl}
    \mA^\top & \text{Transpose of matrix $\mA$}\\
    \mA^+ & \text{Moore-Penrose pseudoinverse of $\mA$}\\
    \mA \odot \mB & \text{Element-wise (Hadamard) product of $\mA$ and $\mB$}\\
    \mA\mB & \text{Matrix product, i.e., $\sum_{k}\mA_{i, k}\mB_{j, k}$}\\
    \mathrm{det}(\mA) & \text{Determinant of $\mA$}\\
\end{array}$

### Calculus

$\begin{array}{ll}
    \frac{d y} {d x} & \text{Derivative of $y$ with respect to $x$}\\
    \frac{\partial y} {\partial x}  & \text{Partial derivative of $y$ with respect to $x$}\\
    \nabla_\vx y  & \text{Gradient of $y$ with respect to $\vx$}\\
    \nabla_\mX y  & \text{Matrix derivatives of $y$ with respect to $\mX$}\\
    \nabla_\tX y  & \text{Tensor containing derivatives of $y$ with respect to $\tX$}\\
    \frac{\partial f}{\partial \vx}  & \text{Jacobian matrix $\mJ \in \sR^{m\times n}$ of $f: \sR^n \rightarrow \sR^m$}\\
    \nabla_\vx^2 f(\vx)\text{ or }\mH( f)(\vx) & \text{The Hessian matrix of $f$ at input point $\vx$}\\
    \int f(\vx) d\vx  & \text{Definite integral over the entire domain of $\vx$}\\
    \int_\sS f(\vx) d\vx & \text{Definite integral with respect to $\vx$ over the set $\sS$}\\
\end{array}$

### Probability and Information Theory

$\begin{array}{cl}
    \ra \bot \rb & \text{The random variables $\ra$ and $\rb$ are independent}\\
    \ra \bot \rb \mid \rc & \text{They are conditionally independent given $\rc$}\\
    P(\ra) & \text{A probability distribution over a discrete variable}\\
    p(\ra) & \text{A probability distribution over a continuous variable, or over}\\
           & \text{a variable whose type has not been specified}\\
    \ra \sim P & \text{Random variable $\ra$ has distribution $P$}\\
    \mathbf{E}_{\rx\sim P} [ f(x) ]\text{ or } \mathbf{E} f(x) & \text{Expectation of $f(x)$ with respect to $P(\rx)$}\\
    \Var(f(x))  &  \text{Variance of $f(x)$ under $P(\rx)$}\\
    \Cov(f(x),g(x))  & \text{Covariance of $f(x)$ and $g(x)$ under $P(\rx)$}\\
    H(\rx)  & \text{Shannon entropy of the random variable $\rx$}\\
    \KL ( P \Vert Q )  & \text{Kullback-Leibler divergence of P and Q}\\
    \mathcal{N} ( \vx ; \vmu , \mSigma) & \text{Gaussian distribution over $\vx$ with
    mean $\vmu$ and covariance $\mSigma$}\\
\end{array}$

### Functions

$\begin{array}{cl}
    f: \sA \rightarrow \sB & \text{The function $f$ with domain $\sA$ and range $\sB$}\\
    f \circ g  & \text{Composition of the functions $f$ and $g$}\\
    f(\vx ; \vtheta)  & \text{A function of $\vx$ parametrized by $\vtheta$.}\\
        & \text{(Sometimes we write $f(\vx)$ and omit the argument $\vtheta$ to lighten notation)}\\
    \log x & \text{Natural logarithm of $x$}\\
    \sigma(x) & \text{Logistic sigmoid, $\displaystyle \frac{1} {1 + \exp(-x)}$}\\
    \zeta(x) & \text{Softplus, $\log(1 + \exp(x))$}\\
    || \vx ||_p  & \text{$L^p$ norm of $\vx$}\\
    || \vx || & \text{$L^2$ norm of $\vx$}\\
    x^+ & \text{Positive part of $x$, i.e., $\max(0,x)$}\\
    1_\mathrm{condition} & \text{is 1 if the condition is true, 0 otherwise}\\
\end{array}$

### Datasets and Distributions

$\begin{array}{cl}
    p_{data} & \text{The data generating distribution}\\
    p_{train} & \text{The empirical distribution defined by the training set}\\
    \sX & \text{A set of training examples}\\
    \vx^{(i)} & \text{The $i$-th example (input) from a dataset}\\
    y^{(i)}\text{ or }\vy^{(i)} & \text{The target associated with $\vx^{(i)}$ for supervised learning}\\
    \mX & \text{The $m \times n$ matrix with input example $\vx^{(i)}$ in row $\mX_{i,:}$}\\
\end{array}$
